```python
#Importing pandas for dataframe
import pandas as pd
```

```python
#Importing the loan tape file. Clean data prior to input. Make sure to take out the zero values in relevant columns.
df = pd.read_csv(r'C:\Users\MADYMA\Desktop\Tale of the Tape\COLT 2021-6 Investor Tape.csv')

#Removing column truncation and 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
#Sanity check
df.head()
```

```python
#Creating CLTV Buckets
bins = [0.20000000001, 0.30, 0.30000000001, 0.40, 0.40000000001, 0.50, 0.50000000001, 0.60, 0.60000000001, 0.70, 0.70000000001, 0.80, 0.80000000001, 0.90, 0.90000000001, 1.00]
df['LTV Buckets'] = pd.cut(df['Original CLTV (%)'], bins)

#Sorting the values by CLTV and resetting the index. Drop will prevent the dataframe from keeping the old index.
LTV_final_df = df.sort_values(by='Original CLTV (%)', ascending=True).reset_index(drop=True)
LTV_final_df.head()
```

```python
#Grouping columns by loan purpose and aggregating different values
loan_purpose_df = LTV_final_df.groupby('Loan Purpose').agg({'% Loan Amount':'sum', 'DTI':'mean', 'Total Income':'mean', 'Liquid Reserve':'mean', 'Original CLTV (%)':'mean', 'FICO':'mean', 
'Length of Employment':'mean', 'IO Term':'count'})
loan_purpose_df
```

```python
#Printing both data frames to an excel workbook
with pd.ExcelWriter(r'C:\Users\MADYMA\Desktop\Tale of the Tape\COLT 2021-6UPDATED.xlsx') as writer: 
    LTV_final_df.to_excel(writer, sheet_name='LTV')
    loan_purpose_df.to_excel(writer, sheet_name='Purpose')
```
