```python
#Importing pandas for dataframe
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\MADYMA\Desktop\Tale of the Tape\JPMMT2022-2InvestorTape.csv')

# Removing column truncation to max visual
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
df.head(11)
```

```python
# Creating CLTV Buckets
bins = [0.10, 0.20000000001, 0.30, 0.30000000001, 0.40, 0.40000000001, 0.50, 0.50000000001, 0.60, 0.60000000001, 0.70, 0.70000000001, 0.80, 0.80000000001, 0.90, 0.90000000001, 1.00]
df['LTV Buckets'] = pd.cut(df['Orign Combined LTV'], bins)

# Sorting the values by CLTV and resetting the index. Drop will prevent the dataframe from keep the old sort values.
LTV_final_df = df.sort_values(by='Orign Combined LTV', ascending=True).reset_index(drop=True)
LTV_final_df.head(49)
```

```python
purpose_df = LTV_final_df.groupby('Loan Purpose').agg({'Back DTI at Orign':'mean', 'Liquid Cash Reserves at Orign':'mean', 'Orign Combined LTV':'mean', 'Orign Combined FICO Score':'mean'})
purpose_df
```

```python
# Printing Loan Purpose Data Frame
with pd.ExcelWriter(r'C:\Users\MADYMA\Desktop\Tale of the Tape\JPMMT 2022-2.xlsx') as writer: 
    LTV_final_df.to_excel(writer, sheet_name='LTV')
    purpose_df.to_excel(writer, sheet_name='Purpose')
```
